```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
file_path = 'output_age.csv'
data = pd.read_csv(file_path)
plt.figure(figsize=(16, 8))

casenum = list(data['agenum'])
deathnum = list(data['deathnum'])
index = list(data['age'])
df = pd.DataFrame({'case number': casenum,
                   'death number': deathnum}, index=index)
ax = df.plot.barh()
plt.title('Case Number and Death Number')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
file_path = 'output_age.csv'
data = pd.read_csv(file_path)
plt.figure(figsize=(16, 8))

bar_data = data[['age', 'mortality']]
plt.plot(bar_data['age'],bar_data['mortality'],'o-',color = 'b')

# Set labels and titles

plt.xlabel('Age group')
plt.ylabel('Infection Fatality Rate')
plt.title('Infection Fatality Rate Over Age')

plt.show()
```
